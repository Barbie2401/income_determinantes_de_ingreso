```python
#definimos la variables no categóricas

variables_no_cat = []

for n, i in enumerate(df_binarizado_new):
    if (i !='income_mayor_50K' and df_binarizado_new[i].dtypes =="int32"):
            variables_no_cat.append(i)
    elif (i !='income_mayor_50K' and df_binarizado_new[i].dtypes =="float64"):
            variables_no_cat.append(i)
    elif (i !='income_mayor_50K' and df_binarizado_new[i].dtypes =="int64"):
            variables_no_cat.append(i)

    elif (i !='income_mayor_50K' and df_binarizado_new[i].dtypes =="uint8"):
            variables_no_cat.append(i)
                
variables_ind = " + ".join(variables_no_cat)

variables_ind
```

Ambos modelos presentan una Pseudo R cuadrado similar, es decir, en ambos casos tiene un valor de **0.408** , esto quiere decir que explica la varianza del modelo un 40%.

Todas estas variables aportan a tener una mayor porbabilidad de presentar un salario mayor a los 50k dolares.

```python
#Coeficientes por variable
coef_variables = modelo_depurado.params
```

```python
#Revisamos los betas
coef_variables
```

```python
#Creamos la funcion inverse_logit

def inverse_logit(x):
    '''
        Dado un objeto numérico, devuelve el logit inverso de los valores.

            Parameters: 
                x: Un objeto numérico. Se permiten valores faltantes (NaN).

            Returns: 
                Un objeto del mismo tipo que x contiene los logit inversos de los valores de entrada
            
            Detalles:
                El logit inverso está definido por exp(x)/(1+exp(x)). 
                Valores en o devuelven logits de 0 o 1 respectivamente x. 
                Cualquier x en la entrada también será x en la salida.

        '''
    return 1 / (1+np.exp(-x))
```

```python
#Log inverso
coef_variables.map(lambda x: round (inverse_logit(modelo_depurado.params['Intercept']+x),4))
```
